# Data Cleaning 

Checking input mushroom balance poisonus to edible ratio

```python
edible_cnt = mushroom[mushroom["class"] == "e"]["class"].count()
poisonus_cnt = mushroom[mushroom["class"] == "p"]["class"].count()
p_e_ratio = poisonus_cnt/float(edible_cnt)
print("\n'poisonus'-to'edible' ratio: {}\npoisonus_cnt: {}, edible_cnt: {}"
      .format(p_e_ratio.round(2),poisonus_cnt,edible_cnt))


import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style="ticks", color_codes=True)
plt.title("Balance Checking for input class in Mushroom Dataset",fontsize=14)
sns.countplot(x = mushroom["class"], data = mushroom)
```

The Mushroom 'Poisonus' to 'Edible' ration is 0.93
in which poisonus count is 3916 and edible count is 4208
