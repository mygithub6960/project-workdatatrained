# Poisonus Or Edible 

```python
edible_cnt=data[data["class"]=="e"]["class"].count()
poisonus_cnt=data[data["class"]=="p"]["class"].count()
p_e_ratio=poisonus_cnt/float(edible_cnt)
print("\n'poisonus'-to-'edible'ratio:{}\npoisonus_cnt:{},edible_cnt:{}".format(p_e_ratio.round(2),poisonus_cnt,edible_cnt))
```

The ratio of poisonus to edible is 0.93, making poisonus count as 3916 and edible count as 4208
